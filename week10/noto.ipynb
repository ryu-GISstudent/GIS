```python
m = folium.Map(location=[36.53129882505882,136.66585425411765
], zoom_start=7)


for index, row in df.iterrows():
    latlon = [row['lat'],row['lon']]
    folium.Marker(latlon, 
                  popup='<div style="width:250px">'+row[['name','remarks','address','phone','fee']]+'</div>', 
                  tooltip=row['name'],
                  icon=folium.Icon(color = 'red')
                ).add_to(m)

m
```
