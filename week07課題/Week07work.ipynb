```python
import geopandas as gpd
import pandas as pd
```

```python
# データを変数に入れる
saitama_ku = gpd.read_file('saitama.geojson')
```

```python
# とりあえずマップとしてアウトプット
saitama_ku.plot()
```

```python
# 最初の５列は？
saitama_ku.sample(6)
```

```python
# データの情報
saitama_ku.info()
```

```python
saitama_pop = pd.read_csv('pp.csv')
```

```python
saitama_pop.head()
```

```python
saitama_pop.info()
```

```python
# rename
saitama_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
saitama_ku
```

```python
saitama_ku['KEY_CODE'] = saitama_ku['KEY_CODE'].astype(int)
```

```python
# checkする
saitama_pop.info()
```

```python
# ここでようやく merge コマンドで統合させて、新しい変数に入れる
saitama1= saitama_ku.merge(saitama_pop,on='KEY_CODE')
```

```python
# check!
saitama1
```

```python
saitama1.info()
```

```python
saitama1['人口総数'] = saitama1['人口総数'].astype (int)
```

```python
# check
saitama1.info()
```

### 4-3 Let's map!

```python
saitama1.plot(column='人口総数',legend=True)
```

```python

saitama1.plot(figsize=(10,10),
                  column='人口総数',
                  legend=True)
```

```python
saitama1.plot(figsize=(10,10),
                  column='人口総数',
                  legend=True,
                  cmap='seismic')
```
