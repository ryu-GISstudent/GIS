```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
ObukuroStation = [35.92418652154995, 139.77793816557548]
ObukuroStationWest7Eleven = [35.92440230851712, 139.7776119398692]
ObukuroStationEast7Eleven = [35.92318701853483, 139.7784683423062]
Simomakuri7Eleven = [35.926068091596136, 139.78238033731785]
大袋セブン = [35.92211703000093, 139.7594470966291]


# make the map
m = folium.Map(location=ObukuroStation, zoom_start=15)

# add a marker
folium.Marker(ObukuroStation, popup="駅です", tooltip="大袋駅", ).add_to(m)
folium.Marker(ObukuroStationWest7Eleven, popup="駅から36M", tooltip="西口店",icon=folium.Icon(color='lightblue')).add_to(m)
folium.Marker(ObukuroStationEast7Eleven, popup="駅から170M", tooltip="東口店",icon=folium.Icon(color='red')).add_to(m)
folium.Marker(Simomakuri7Eleven, popup="駅から550M", tooltip="下間久里店",icon=folium.Icon(color='red')).add_to(m)
folium.Marker(大袋セブン, popup="駅から1.7KM", tooltip="大袋店", icon=folium.Icon(color='lightblue')).add_to(m)

folium.CircleMarker([35.92418652154995, 139.77793816557548],
                radius=30,
                popup='Laurelhurst Park',
                color='#3186cc',
                fill_color='#3186cc',
               ).add_to(m)

# show the map
m.save("map.html")

m
```
